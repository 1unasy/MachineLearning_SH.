# ◎ 판다스 개요와 기본 API

### 판다스 주요 구성 요소
- DataFrame : Column X Rows **2차원** 데이터셋
    
    
- Series : 1개의 **Column 값**으로만 구성된 **1차원** 데이터셋
    
    
- Index : DataFrame/Series의 고유한 **Key값** 객체



- 출처 : https://www.w3resource.com/python-exercises/pandas/index.php

```python
import pandas as pd
```

#### < 데이터 로드 >

#### read_csv()
- read_csv()를 이용하여 csv 파일을 편리하게 DataFrame으로 로딩한다.
- read_csv()의 sep 인자로 콤마(,)가 아닌 다른 분리자를 사용해 파일을 로드할 수 있다.

```python
titanic_df = pd.read_csv('titanic_train.csv')
type(titanic_df) # 데이터프레임(DataFrame)
```

#### head()
- DataFrame의 맨 앞 일부 데이터만 추출한다.
- Default == 5

```python
titanic_df.head(3)
```

#### < DataFrame 생성 >

```python
dic1 = {'Name': ['Chulmin', 'Eunkyung','Jinwoong','Soobeom'],
        'Year': [2011, 2016, 2015, 2015],
        'Gender': ['Male', 'Female', 'Male', 'Male']
       }
# 딕셔너리를 DataFrame으로 변환
# key : 컬럼명, list : Series/DataFrame의 values
data_df = pd.DataFrame(dic1)
print(data_df)
print("#"*30)

# 새로운 컬럼명을 추가
data_df = pd.DataFrame(dic1, columns=["Name", "Year", "Gender", "Age"])
print(data_df)
print("#"*30)

# 인덱스를 새로운 값으로 할당. 
data_df = pd.DataFrame(dic1, index=['학생1','학생2','학생3','학생4'])
print(data_df)
print("#"*30)
```

#### < DataFrame의 컬럼명과 인덱스 >

```python
print("columns:",titanic_df.columns) # 컬럼명
print("index:",titanic_df.index) # 인덱스
print("index value:", titanic_df.index.values) # 데이터 값
```

#### < DataFrame에서 Series 추출 및 DataFrame 필터링 추출 >

```python
# DataFrame객체에서 []연산자내에 한개의 컬럼만 입력하면 Series 객체를 반환  
series = titanic_df['Name']
print(series.head(3))
print("## type:",type(series))
print()

# DataFrame객체에서 []연산자내에 여러개의 컬럼을 리스트로 입력하면 그 컬럼들로 구성된 DataFrame 반환  
filtered_df = titanic_df[['Name', 'Age']]
print(filtered_df.head(3))
print("## type:", type(filtered_df))
print()

# DataFrame객체에서 []연산자내에 한개의 컬럼을 리스트로 입력하면 한개의 컬럼으로 구성된 DataFrame 반환 
one_col_df = titanic_df[['Name']]
print(one_col_df.head(3))
print("## type:", type(one_col_df))
```

#### shape 
- DataFrame의 행(Row)와 열(Column) 크기를 가지고 있는 속성이다.

```python
print('DataFrame 크기: ', titanic_df.shape) # (행의 갯수, 열의 갯수)
```

#### info()
- DataFrame내의 컬럼명, 데이터 타입, Null건수, 데이터 건수 정보를 제공한다.

```python
titanic_df.info()
```

#### describe()
- 데이터값들의 평균,표준편차,4분위 분포도를 제공한다. 
- 즉, **숫자형 컬럼들에 대한 정보**를 제공한다.

```python
titanic_df.describe()
```

#### value_counts()  
- 동일한 개별 데이터 값이 몇건이 있는지 정보를 제공한다. 
- 즉 **개별 데이터값의 분포도**를 알려준다. 
- 주의할 점: value_counts()는 Series객체에서만 호출 될 수 있으므로 반드시 DataFrame을 단일 컬럼으로 입력하여 **Series로 변환한 뒤 호출**한다. 

```python
value_counts = titanic_df['Pclass'].value_counts()
print(value_counts)
```

```python
titanic_pclass = titanic_df['Pclass']
print(type(titanic_pclass))
```

```python
titanic_pclass.head()
```

```python
value_counts = titanic_df['Pclass'].value_counts()
print(type(value_counts))
print(value_counts)
```

#### sort_values()
- by=정렬컬럼, ascending=True 또는 False로 오름차순/내림차순으로 정렬

```python
titanic_df.sort_values(by='Pclass', ascending=False) # 전체 데이터프레임에 대해서 Pclass를 기준으로 오름차순 정렬
```

```python
titanic_df[['Name','Age']].sort_values(by='Age') # 데이터프레임 중 Name, Age 컬럼을 오름차순 정렬
```

```python
# 데이터프레임 중 Name, Age, Pclass 컬럼을 Pclass, Age 순서대로 오름차순 정렬
titanic_df[['Name','Age','Pclass']].sort_values(by=['Pclass','Age']) 
```
